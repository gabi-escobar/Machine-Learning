```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('Impact_of_Remote_Work_on_Mental_Health.csv')
df
```

```python
# Datos nulos
df.isna().sum()
```

```python
# Crear un DataFrame de ejemplo con los datos mencionados
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hank', 'Ivy', 'Jack'],
    'Age': [18, 19, 18, 20, 17, 19, 21, 22, 18, 20],
    'Gender': ['F', 'M', 'M', 'M', 'F', 'M', 'F', 'M', 'F', 'M'],
    'Admission Test Score': [85, 90, 88, 92, 78, 83, 87, 91, 84, 89],
    'High School Percentage': [90, 88, 89, 93, 85, 86, 91, 92, 88, 90],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose'],
    'Admission Status': ['Admitted', 'Not Admitted', 'Admitted', 'Admitted', 'Not Admitted', 'Admitted', 'Admitted', 'Not Admitted', 'Admitted', 'Admitted']
}
df = pd.DataFrame(data)

# Mostrar datos originales
print("Datos originales:")
print(df)

# Eliminar valores nulos
df_cleaned = df.dropna()

# Verificar que no hay valores nulos restantes
print("\nValores nulos después de la limpieza:")
print(df_cleaned.isna().sum())

# Asegurarse de que no hay datos duplicados
df_cleaned = df_cleaned.drop_duplicates()

# Mostrar datos limpios
print("\nDatos limpios:")
print(df_cleaned)

# Descripción básica
print("\nDescripción básica del DataFrame limpio:")
print(df_cleaned.describe())
print(df_cleaned.info())
```

```python
#visualizacion

# Análisis Univariado
plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['Age'], bins=5)
plt.title('Distribución de Edad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x='Gender', data=df_cleaned)
plt.title('Distribución por Género')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['Admission Test Score'], bins=5)
plt.title('Distribución del Puntaje del Examen de Admisión')
plt.xlabel('Puntaje del Examen de Admisión')
plt.ylabel('Frecuencia')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['High School Percentage'], bins=5)
plt.title('Distribución del Porcentaje de Calificaciones de la Escuela Secundaria')
plt.xlabel('Porcentaje de Calificaciones de la Escuela Secundaria')
plt.ylabel('Frecuencia')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x='Admission Status', data=df_cleaned)
plt.title('Estado de Admisión')
plt.show()
```

```python
#correlacion

# Matriz de Correlación
numeric_columns = df_cleaned.select_dtypes(include=['int64', 'float64']).columns.tolist()
correlation_matrix = df_cleaned[numeric_columns].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis')
plt.title('Matriz de Correlación entre Variables Numéricas')
plt.show()

# Identificación de Outliers
def identify_outliers_iqr(df, columns):
    outlier_indices = []
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1

        # Definición de límites
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Identificación de outliers
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        outlier_indices.extend(outliers.index.tolist())

    return set(outlier_indices)

# Identificar outliers en las variables numéricas
outlier_indices = identify_outliers_iqr(df_cleaned, numeric_columns)
num_outliers = len(outlier_indices)
num_data = len(df_cleaned)

print(f"Número total de outliers identificados: {num_outliers}")
print(f"Porcentaje de outliers: {(num_outliers / num_data) * 100:.2f}%")
```

```python
import pandas as pd

# Cargar el CSV con low_memory=False
df = pd.read_csv('StudentGradesAndPrograms.csv', low_memory=False)

# Verificar los nombres de las columnas
print("Nombres de las columnas del DataFrame:")
print(df.columns)

# Convertir los tipos de datos después de la carga
df['gradeLevel'] = pd.to_numeric(df['gradeLevel'], errors='coerce')
df['gradePercentage'] = pd.to_numeric(df['gradePercentage'], errors='coerce')

# Mostrar datos con tipos de datos convertidos
print("\nDatos con tipos de datos convertidos:")
print(df.head())
print(df.info())
```

```python
# Verificar los nombres de las columnas
print("Nombres de las columnas del DataFrame:")
print(df.columns)

# Verificar los valores únicos en la columna 'gradeLevel' para identificar problemas
print("Valores únicos en 'gradeLevel':")
print(df['gradeLevel'].unique())

# Verificar los valores únicos en la columna 'gradePercentage' para identificar problemas
print("Valores únicos en 'gradePercentage':")
print(df['gradePercentage'].unique())

# Convertir los tipos de datos después de la carga
df['gradeLevel'] = pd.to_numeric(df['gradeLevel'], errors='coerce')
df['gradePercentage'] = pd.to_numeric(df['gradePercentage'], errors='coerce')

# Eliminar filas con valores no numéricos convertidos en NaN
df_cleaned = df.dropna(subset=['gradeLevel', 'gradePercentage'])

# Verificar que no hay valores nulos restantes
print("\nValores nulos después de la limpieza:")
print(df_cleaned.isna().sum())

# Asegurarse de que no hay datos duplicados
df_cleaned = df_cleaned.drop_duplicates()

# Mostrar datos limpios
print("\nDatos limpios:")
print(df_cleaned.head())

# Descripción básica del DataFrame limpio
print("\nDescripción básica del DataFrame limpio:")
print(df_cleaned.describe())
print(df_cleaned.info())
```
